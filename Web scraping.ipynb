```python
import tkinter
from tkinter import *
from tkinter.ttk import *


root = tkinter.Tk()

root.geometry("800x600")
root.configure(bg='lavender')

root.title("INDUSTRY NEWS ANALYSIS")

lab=tkinter.Label(root,text="CHECK OUT THE LATEST INDUSTRY NEWS IN INDIA!",bg='pink',font=('Arial',20,'bold','underline'))
lab.pack(pady=20)
label = tkinter.Label(root, text="ENTER  THE  NEWS  TOPIC",bg='lavender',font=('Times New Roman',20,'bold'))
label.pack(pady=20)
entry = tkinter.Entry(root,bd=6)
entry.pack(pady=10)


def on_button():
    
    url = "https://economictimes.indiatimes.com/news/politics"
    a = Article(url, language="en")
    a.download()
    a.parse()
    a.nlp()
    
    url2 = "https://economictimes.indiatimes.com/industry"
    b = Article(url2, language="en")
    b.download()
    b.parse()
    b.nlp()
    
    url3 = "https://www.ibef.org/industry/indian-consumer-market.aspx"
    c = Article(url3, language="en") 
    c.download()
    c.parse()
    c.nlp()
    
    url4 = "https://economictimes.indiatimes.com/small-biz"
    d = Article(url4, language="en") 
    d.download()
    d.parse()
    d.nlp()
    
    url5 = "https://economictimes.indiatimes.com/mf/mf-news"
    e = Article(url5, language="en") 
    e.download()
    e.parse()
    e.nlp()
    
    url6 = "https://economictimes.indiatimes.com/tech"
    f = Article(url6, language="en")
    f.download()
    f.parse()
    f.nlp()
    
    url7 = "https://economictimes.indiatimes.com/jobs"
    g = Article(url7, language="en")
    g.download()
    g.parse()
    g.nlp()
    
    url8 = "https://economictimes.indiatimes.com/nri"
    h = Article(url8, language="en")
    h.download()
    h.parse()
    h.nlp() 
    
    
    
    if entry.get() == "politics":
        slabel = tkinter.Label(root, text=a.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20,)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, a.text)
        
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
        
    elif entry.get() == "industry":
        slabel = tkinter.Label(root, text=b.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, b.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    elif entry.get() == "markets":
        slabel = tkinter.Label(root, text=c.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 1000)
        T.pack()
        T.insert(tkinter.END, c.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    elif entry.get() == "RISE":
        slabel = tkinter.Label(root, text=d.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, d.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    elif entry.get() == "mutualfunds":
        slabel = tkinter.Label(root, text=e.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, e.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    elif entry.get() == "technology":
        slabel = tkinter.Label(root, text=f.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, f.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    elif entry.get() == "jobs":
        slabel = tkinter.Label(root, text=g.title, font=('Arial', 16 ))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, g.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)

    elif entry.get() == "NRI":
        slabel = tkinter.Label(root, text=h.title, font=('Arial', 16))
        slabel.pack(fill=tkinter.X, pady=20)
        T = tkinter.Text(root, height = 300, width = 800)
        T.pack()
        T.insert(tkinter.END, h.text)
        Font_tuple = ("Times New Roman", 14)
        T.configure(font = Font_tuple)
        
    else:
        tlabel = tkinter.Label(root, text="")
        tlabel.pack(fill=tkinter.X, pady=20)
        slabel1 = tkinter.Label(root, text=g.text, font=('Arial', 12 ))
        slabel1.pack()


photo = PhotoImage(file = r"Capture.png")
photoimage = photo.subsample(8, 8)
button = tkinter.Button(root, text="ENTER", image = photoimage,compound = LEFT, command=on_button,bg='pink',bd=3)
button.pack()


root.mainloop()
```
